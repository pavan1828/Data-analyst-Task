## For each month - how many Data Bank customers make more than 1 deposit and either 1 purchase or 1 withdrawal in a single month?


```python
cus_t['txn_date'] = pd.to_datetime(cus_t['txn_date'])
cus_t['month'] = cus_t['txn_date'].dt.to_period('M')
```

```python

transaction_count_per_month = cus_t.groupby(['customer_id', cus_t['txn_date'].dt.to_period('M')]).agg(
    deposit_count=('txn_type', lambda x: (x == 'deposit').sum()),
    withdrawal_count=('txn_type', lambda x: (x == 'withdrawal').sum()),
    purchase_count=('txn_type', lambda x: (x == 'purchase').sum())
).reset_index()

# Filter customers who made more than 1 deposit and either 1 purchase or 1 withdrawal in a single month
filtered_data = transaction_count_per_month[(transaction_count_per_month['deposit_count'] > 1) &
                                            ((transaction_count_per_month['purchase_count'] == 1) |
                                             (transaction_count_per_month['withdrawal_count'] == 1))]

# Count customers for each month
result_df = filtered_data.groupby('txn_date').size().reset_index(name='customer_count')

# Display the result DataFrame
result_df
```
